# 3ª Tarefa

**Para cada indicação abaixo, construa um novo dataframe, salvando-o com a terminação .csv.**

**(a) ranking de acidentes por estado para ambos os anos (um só arquivo combinando a
informação dos dois anos);**

```python
acidente2021 = df2021['uf'].value_counts()
dados = {'2021':acidente2021}
acidentes = pd.DataFrame(dados).rename_axis('UF').reset_index().sort_values(by = ['2021']).loc[::-1,:]
acidentes.to_csv('ranking2021.csv',sep=';',index = None, header=True)
acidentes
```

**(b) acidentes por dia da semana para ambos os anos (um só arquivo combinando a informação dos dois anos).**

```python
semana2021 = df2021['dia_semana'].value_counts()
dados = {'2021':semana2021}
crshwk = pd.DataFrame(dados).rename_axis('dia').reset_index().sort_values(by = ['2021']).loc[::-1,:]
crshwk.to_csv('acidentes-semana-21.csv',sep=';',index = None, header=True)
crshwk
```
